```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import xgboost as xgb
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from xgboost import XGBClassifier

# Carregar os datasets
df1 = pd.read_csv("/winequality-red.csv")
df1['type'] = '0'
df2 = pd.read_csv("/winequality-white.csv")
df2['type'] = '1'
weather_df = pd.read_csv("/weather-porto-2024.csv", delimiter=';')

# Combinar os datasets
wine_df = pd.concat([df1, df2], ignore_index=True)
wine_df.columns = wine_df.columns.str.replace(' ', '_')

# Filtrar estao de vero (Agosto, Setembro e Outubro)
weather_df['timestamp'] = pd.to_datetime(weather_df['timestamp'], format='%d/%m/%Y %H:%M')
weather_filtered = weather_df[(weather_df['timestamp'].dt.month >= 8) & (weather_df['timestamp'].dt.month <= 10)]

# Gerar datas aleatrias dentro do intervalo do dataset de clima
min_date = weather_filtered['timestamp'].min()
max_date = weather_filtered['timestamp'].max()
wine_df['timestamp'] = np.random.choice(pd.date_range(min_date, max_date), size=len(wine_df))

# Mesclar datasets com base no timestamp
df = pd.merge_asof(wine_df.sort_values('timestamp'),
                    weather_df.sort_values('timestamp'),
                    on='timestamp',
                    direction='nearest')

# Remover colunas no numricas e irrelevantes
df.drop(columns=['timestamp', 'entity_id', 'entity_type', 'name', 'barometricpressure', 'uv_index', 'date_observed', 'time_observed'], inplace=True, errors='ignore')

# Garantir que todas as colunas so numricas
df = df.apply(pd.to_numeric, errors='coerce')

# Criar feature target baseada em quality e reindexar os valores
min_quality = df['quality'].min()
df['quality'] = df['quality'] - min_quality
target = df['quality']
features = df.drop(['quality'], axis=1)

# Diviso do conjunto de dados
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=40)

# Resetar os ndices
xtest = pd.DataFrame(xtest, columns=features.columns).reset_index(drop=True)
ytest = ytest.reset_index(drop=True)

# Tratamento de valores ausentes
imputer = SimpleImputer(strategy='mean')
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)

# Normalizao
scaler = MinMaxScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)

# Definio dos modelos
models = [
    LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs'),
    XGBClassifier(objective='multi:softmax', num_class=len(target.unique())),
    SVC(kernel='rbf', probability=True),
    RandomForestClassifier(n_estimators=100, random_state=42),
    GradientBoostingClassifier(n_estimators=100, random_state=42)
]

# Funo de avaliao dos modelos
def evaluate_model(model, xtrain, ytrain, xtest, ytest):
    model.fit(xtrain, ytrain)
    ytest_pred = model.predict(xtest)
    print(f'Validation Accuracy: {model.score(xtest, ytest):.4f}')
    print(classification_report(ytest, ytest_pred))
    cm = confusion_matrix(ytest, ytest_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.show()

for model in models:
    print(f'\nEvaluando {model.__class__.__name__}:')
    evaluate_model(model, xtrain, ytrain, xtest, ytest)

# Treinar o melhor modelo e salvar
best_model = models[1]
best_model.fit(xtrain, ytrain)
joblib.dump(best_model, 'wine_quality_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(imputer, 'imputer.pkl')
```
